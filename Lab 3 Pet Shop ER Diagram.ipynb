Lab 3: Pet Shop ER Diagram

   As we are selling animals (pets and fish) and food, there must be some customers who are willing to buy them, which is why we have created an entity of customer. Purchase is another entity and we have split this up into two different entities, animal purchase and food purchase, which will allow for us to track them separately. Finally, food, fish and pets are also the main entities because we need to track which one in sold. Additionally, loyalty program will be an optional part and we will create this entity.  We have also added some attributes under each entity, including foreign keys. For customer entity, customer id, first name, last name, etc. will be on the table, so basic information that will help us track the customer. For purchase entity, purchase ID, customer ID, animals and food ID (separately) are tracked. Recall, if the primary key of an entity is in another entity, it will be the foreign key on the bottom of the table. Fish, pets and food are our goods. Fish are kept in aquariums and pets are kept in enclosures. So we can easily manage our animal storage. So aquarium ID and enclosure ID will be on each table. Other attributes will be ID, type, etc. (basic info). 
   
   After creating all the entities that we need, we will find the relationships between them. Relationship from customer to purchase will be zero-to-many because a customer can buy either nothing or everything. Relationship from purchase to customer will be one-to-one only because a specific purchase will be generated by a specific customer. Relationship from purchase to (fish, pets and food) will be one-to-many because if the purchase is made, there must be at least one item purchased. Relationship from (fish, pets and food) will be zero-to-many because an item does not need to be in the purchase and items in each category can be in different many purchases. Relationship from aquariums and enclosures to fish and pets (kept locations to animals) will be one-and-one-only because a specific animal is in a specific location. The same animal cannot be in different locations. Relationship from animals to locations will be zero-to-one because a specific location category will hold a specific type of animals or nothing which means one or zero. Relationship between loyalty program and the customer will be one-and-one-only and one-and-one-only on the other hand because a specific customer will have a specific loyalty program account.
